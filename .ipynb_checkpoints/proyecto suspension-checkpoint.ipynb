### Objetivo específico:

Simular el modelo de una suspensión de un automóvil, representada en una cuarta parte de la misma.

Las ecuaciones que modelan el problema son: 
>- zr = A*np.sin(w*t)
>- dzr = w*A*np.cos(w*t)
    

## <font color=blue> Simulaciones

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.integrate import odeint
```

```python
Kwr = 500000
Kcw = 80000
Ccw = 350
Cwr = 15020
mc = 2500
mw = 320
def suspen(x, t):
    x1 = x[0]
    x2 = x[1]
    x3 = x[2]
    x4 = x[3]
    #zr, dzr = 0, 0
    A, w = 0.05, 50
    zr = A*np.sin(w*t)
    dzr = w*A*np.cos(w*t)
    return [x2, -Kcw/mc * (x1-x3) - Ccw/mc * (x2-x4), x4, Kcw/mw * (x1-x3) + Ccw/mw * (x2-x4) - Kwr/mw * (x3-zr) - Cwr/mw * (x4-dzr)]
```

```python
#Condición Inicial
x0 = [0, 0, 0, 0]
tt = np.linspace(0, 20, 100)

xx = odeint(suspen, x0, tt)

x1 = xx[:, 0]
x2 = xx[:, 1]
x3 = xx[:, 2]
x4 = xx[:, 3]

plt.figure(figsize=(10,5))

plt.plot(tt, x1, '-b', label= '$x_1=$ Posición del carro')
#plt.plot (tt, x2, '--r', label = '$x_2=$')
#plt.plot(tt, x3, '--k', label= '$x_3=$')
#plt.plot (tt, x4, '#04B404', label = '$x_4=$')
plt.grid()
plt.legend(loc='best')
plt.xlabel('Tiempo')
plt.ylabel('Posición')

plt.show()
```

```python
plt.plot(tt, x1, '-b', label= '$x_1=$ Posición del carro')
plt.grid()
plt.legend(loc='center',  bbox_to_anchor=(1.25,0.5))
plt.xlabel('Tiempo')
plt.ylabel('Posición')

plt.show()
```

```python
plt.plot (tt, x2, '--r', label = '$x_2=$ Velocidad del carro')
plt.grid()
plt.legend(loc='center',  bbox_to_anchor=(1.25,0.5))
plt.xlabel('Tiempo')
plt.ylabel('Velocidad')

plt.show()
```

```python
plt.plot(tt, x3, '--k', label= '$x_3=$ Posición de la rueda')
plt.grid()
plt.legend(loc='center',  bbox_to_anchor=(1.27,0.5))
plt.xlabel('Tiempo')
plt.ylabel('Posición')

plt.show()
```
